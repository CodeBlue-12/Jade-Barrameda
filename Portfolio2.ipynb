```python
import math #Value of pi is not defined without this import

#Classifying Modes
print ('Welcome to the coded conversion calculator!')
print ('1 - Temperature')
print ('2 - Angles')
print ('3 - Metric System')
print ('Select Mode:')
x=int(input()) 
if x == 1:
        #Classifying Modes
        print ('Select Mode')
        print ('1 - Fahrenheit to Celcius')
        print ('2 - Celcius to Fahrenheit')
        print ('3 - Celcius to Kelvin')
        print ('4 - Fahrenheit to Kelvin')
        print ('5 - Kelvin to Celcius')
        print ('6 - Kelvin to Fahrenheit')
        a=int(input('Mode:'))

        if a == 1:
            print ('Running mode - Fahrenheit to Celcius')
            y=float(input ('Input value in Fahrenheit:'))
            z=(y - 32) * (5/9) #Conversion Factor
            print ( y , '°F =' , round(z,2) , '°C')

        elif a == 2:
            print ('Running mode - Celcius to Fahrenheit')
            y=float(input ('Input value in Celcius:'))
            z=(y * (9/5)) + 32 #Conversion Factor
            print ( y , '°C =' , round(z,2) , '°F')

        elif a == 3:
            print ('Running mode - Celcius to Kelvin')
            y=float(input ('Input value in Celcius:'))
            z= y + 273.15 #Conversion Factor
            print ( y , '°C =' , round(z,2) , 'K')

        elif a == 4:
            print ('Running mode - Fahrenheit to Kelvin')
            y=float(input ('Input value in Fahrenheit:'))
            z=((y - 32) * (5/9)) + 273.15 #Conversion Factor
            print ( y , '°F =' , round(z,2) , 'K')

        elif a == 5:
            print ('Running mode - Kelvin to Celcius')
            y=float(input ('Input value in Kelvin:'))
            z= y - 273.15 #Conversion Factor
            print ( y , 'K =' , round(z,2) , '°C')

        elif a == 6:
            print ('Running mode - Kelvin to Fahrenheit')
            y=float(input ('Input value in Kelvin:'))
            z=((y - 273.15) * (9/5)) + 32 #Conversion Factor
            print ( y , 'K =' , round(z,2) , '°F')
        else:
            print ('Error, try again')

elif x == 2:
    #Classifying Modes
    print ('Select Mode')
    print ('1 - Degrees to Radian')
    print ('2 - Radian to Degrees')
    print ('3 - Degrees to Gradian')
    print ('4 - Gradian to Degrees')
    print ('5 - Radian to Gradian')
    print ('6 - Gradian to Radian')
    x=int(input('Mode:'))

    if x == 1:
            print ('Running mode - Degrees to Radian')
            y=float(input ('Input value in Degrees:'))
            z= y * (math.pi / 180) #Conversion Factor
            print ( y , '°', '=', round(z,4), 'rad')

    elif x == 2:
            print ('Running mode - Radian to Degrees')
            y=float(input ('Input value in Radian:'))
            z= y * (180 / math.pi) #Conversion Factor
            print ( y , 'rad', '=', round(z,4), '°')

    elif x == 3:
            print ('Running mode - Degrees to Gradian')
            y=float(input ('Input value in Degrees:'))
            z= y * (200/ 180) #Conversion Factor
            print ( y , '°', '=', round(z,4), 'g')

    elif x == 4:
            print ('Running mode - Gradian to Degrees')
            y=float(input ('Input value in Gradian:'))
            z= y * (180 / 200) #Conversion Factor
            print ( y , 'g', '=', round(z,4), '°')

    elif x == 5:
            print ('Running mode - Radian to Gradian')
            y=float(input ('Input value in Radian:'))
            z= y * (200/ math.pi) #Conversion Factor
            print ( y , 'rad', '=', round(z,4), 'g')

    elif x == 6:
            print ('Running mode - Gradian to Radian')
            y=float(input ('Input value in Gradian:'))
            z= y * (math.pi / 200) #Conversion Factor
            print ( y , 'g', '=', round(z,4), 'rad')
    else:
        print ('error')

elif x == 3:
    print ('Select unit:')
    print ('g - Grams')
    print ('m - Meters')
    print ('L - Liters')
    a = str(input('Unit:')) #Units for output

    #Input
    print ('Select prefix (from):')
    print ('0.001 - milli')
    print ('0.01 - centi')
    print ('0.1 - deci')
    print ('1 - standard')
    print ('10 - deca')
    print ('100 - hecto')
    print ('1000 - kilo')
    b = float(input('Prefix:'))
    if b == 0.001:
        prefix = 'm'
    elif b == 0.01:
        prefix = 'c'
    elif b == 0.1:
        prefix = 'd'
    elif b == 1:
        prefix = ' ' 
    elif b == 10:
        prefix = 'da'
    elif b == 100:
        prefix = 'h'
    elif b == 1000:
        prefix = 'k'
    else:
        print ('error')

    #Output
    print ('Select prefix (to):')
    print ('0.001 - milli')
    print ('0.01 - centi')
    print ('0.1 - deci')
    print ('1 - standard')
    print ('10 - deca')
    print ('100 - hecto')
    print ('1000 - kilo')
    c = float(input('Prefix:'))
    if c == 0.001:
        prefix1 = 'm'
    elif c == 0.01:
        prefix1 = 'c'
    elif c == 0.1:
        prefix1 = 'd'
    elif c == 1:
        prefix1 = ' ' 
    elif c == 10:
        prefix1 = 'da'
    elif c == 100:
        prefix1 = 'h'
    elif c == 1000:
        prefix1 = 'k'
    else:
        print ('error')

    #Initial Value
    w = float(input ('Insert initial value:'))
    #Processing
    f = w*(b/c)
    #Output
    print ( w, prefix + a, '=', f, prefix1 + a)
    
else:
    print ('Error 404') #Invalid Input
    
```

```python
#Just for fun
import random #This imports a random number generator which will be utilized in the program

print ('Welcome to the coded "Slot Machine Game"!!!')
input ('If you wish to start, insert coin:') #Just something to pause the prompt

#Generation of random values
x = random.randrange(0,9)
y = random.randrange(0,9)
z = random.randrange(0,9)
print ('[',x,y,z,']')

#Conditionals and outputs
if x==y==z:
    print ('Jackpot!!! You sure are lucky!')
elif x==y and x!=z:
    print ('A pair! So close, try your luck one more time!')
elif x==z and x!=y:  
    print ('A pair! So close, try your luck one more time!')
elif y==z and y!=x:
    print ('A pair! So close, try your luck one more time!')
else:
    print ('A dud... Better luck next time!')
```
