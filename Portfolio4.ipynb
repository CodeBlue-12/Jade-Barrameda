```python
# To show current date and time
import datetime
now = datetime.datetime.now()

#To insert delays in-between the prompts
import time

# User Interface
print("Booting up...")
time.sleep(2.5)
print('Welcome to the coded Budget Tracker!')
time.sleep(1)
print('To continue, answer the following questions:')

# Name confirmation
while True:
    name = input('What is your name? ')
    print('Your name is', name, 'correct? (Press Y to continue)')
    if input().upper() == 'Y':
        break

# Age Verification
while True:
    age = input('How old are you? ')
    print('Your age is', age, 'correct? (Press Y to continue)')
    if input().upper() == 'Y':
        break

# Setting up the Initial Values
print('Welcome,', name)
print('Before we begin, input initial values for your cash and card:')
total_cash = float(input('Cash: '))
total_credit = float(input('Card: '))
total_expenditure = 0

# Continuous Mode Selection
while True:
    print('\nSelect mode:')
    print('1 - Income')
    print('2 - Expenditures')
    print('3 - Transaction')
    print('4 - Receipt')
    print('Other - Exit')
    mode = input ()

    if mode == 1:
        print('Select Account:')
        print('1 - Cash')
        print('2 - Card')
        try:
            account = int(input())
        except ValueError:
            print("Please enter a valid number.")
            continue

        if account == 1:
            try:
                cash_amount = float(input('Amount earned (cash): '))
            except ValueError:
                print("Please enter a valid number.")
                continue
            detail = input('Description: ')
            total_cash += cash_amount
            print('Total Income (Cash): Php', total_cash)

        elif account == 2:
            try:
                card_amount = float(input('Amount earned (card): '))
            except ValueError:
                print("Please enter a valid number.")
                continue
            detail = input('Description: ')
            total_credit += card_amount
            print('Total Income (Card): Php', total_credit)

        else:
            print('ERROR 404')

    elif mode == 2:
        while True:
            try:
                expense = float(input('Amount expended: '))
            except ValueError:
                print("Please enter a valid number.")
                continue
            if expense > total_cash:
                print("Expense exceeds available cash. Try again.")
                continue
            detail = input('Description: ')
            total_expenditure += expense
            total_cash -= expense
            print('Total Expenses: Php', total_expenditure)
            add = input('Do you wish to add more? (Press Y to add) ')
            if add != 'Y':
                break

    elif mode == 3:
        print('Select Transaction:')
        print('1 - Cash to Card')
        print('2 - Card to Cash')
        try:
            t = int(input())
        except ValueError:
            print(" Please enter a valid number.")
            continue

        if t == 1:
            try:
                deposit = float(input('Deposit amount: '))
            except ValueError:
                print(" Please enter a valid number.")
                continue
            if deposit > total_cash:
                print(' Insufficient cash.')
                continue
            total_cash -= deposit
            total_credit += deposit
            print('Cash to Card complete.')

        elif t == 2:
            try:
                withdraw = float(input('Withdraw amount: '))
            except ValueError:
                print("Please enter a valid number.")
                continue
            if withdraw > total_credit:
                print('Insufficient credit.')
                continue
            total_credit -= withdraw
            total_cash += withdraw
            print('Card to Cash complete.')

    elif mode == 4:
        print("*" * 40)
        print("          SUMMARY OF INPUT          ")
        print(now)
        print("*" * 40)
        print('Total cash earned: Php', total_cash)
        print('Total card earned: Php', total_credit)
        print('Total expenditures: Php', total_expenditure)
        print('Total cash remaining: Php', total_cash - total_expenditure)

    else:
        print('Thank you for testing!')
        break
```
