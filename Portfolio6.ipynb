```python
def gwa_calculator():
    print("=============================================")
    print("      GWA & ACADEMIC HONORS CALCULATOR       ")
    print("=============================================")
    print("Criteria:")
    print("- President's List: GWA of 1.45 or higher (1.00 to 1.45)")
    print("- Dean's List:      GWA of 1.75 or higher (1.46 to 1.75)")
    print("- Disqualification: Any individual grade lower than 2.40 (e.g., 2.50)")
    print("=============================================")
    
    # Initial empty lists
    subject_names = []
    subject_units = []
    subject_grades = []
    
    while True:
        try:
            num_subjects = int(input("How many graded subjects do you want to enter? "))
            if num_subjects > 0:
                break
            print("Please enter a number greater than 0.")
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    # Data entry loop: Prompts for the course details before collecting the grade
    for i in range(num_subjects):
        print("\n--- Subject " + str(i + 1) + " ---")
        name = input("Enter Subject/Course Name: ").strip()
        
        while True:
            try:
                units = float(input("Enter Units/Credits for " + name + ": "))
                if units > 0:
                    break
                print("Units must be greater than 0.")
            except ValueError:
                print("Invalid input. Please enter a valid number for units.")
                
        while True:
            try:
                grade = float(input("Enter Grade received for " + name + " (1.00 - 5.00): "))
                if 1.0 <= grade <= 5.0:
                    break
                print("Grade is invalid. Please enter a grade between 1.00 and 5.00.")
            except ValueError:
                print("Invalid input. Please enter a numeric grade.")
        
        # Adding values directly to their respective lists
        subject_names.append(name)
        subject_units.append(units)
        subject_grades.append(grade)

    # Calculation of GWA 
    total_weighted_grades = 0.0
    total_units = 0.0
    has_failing_grade = False 
    
    for i in range(len(subject_grades)):
        total_weighted_grades += subject_grades[i] * subject_units[i]
        total_units += subject_units[i]
        if subject_grades[i] > 2.40:  
            has_failing_grade = True
        
    current_gwa = total_weighted_grades / total_units

    # Simple Grade Report Summary Output
    print("\n=============================================")
    print("               GRADE REPORT                 ")
    print("=============================================")
    print("Subject/Course       | Units  | Grade")
    print("---------------------------------------------")
    for i in range(len(subject_names)):
        spaced_name = subject_names[i] + " " * (20 - len(subject_names[i]))
        print(spaced_name + " | " + str(subject_units[i]) + "    | " + str(subject_grades[i]))
    print("---------------------------------------------")
    print("Total Units Evaluated: " + str(total_units))
    print("Calculated GWA:        " + str(round(current_gwa, 2)))
    print("=============================================")

    # Qualification validation checks
    is_president_lister = False
    is_dean_lister = False

    if has_failing_grade:
        print("Status: Not Qualified for Honors")
        print("Reason: You have an individual subject grade that dropped below 2.40.")
    elif current_gwa <= 1.45:
        print("STATUS: CONGRATULATIONS! You qualify for the PRESIDENT'S LIST!")
        is_president_lister = True
    elif current_gwa <= 1.75:
        print("STATUS: CONGRATULATIONS! You qualify for the DEAN'S LIST!")
        is_dean_lister = True
    else:
        print("Status: Your GWA does not meet the minimum honors threshold (1.75).")
    
    # Missing Grade Forecast System
    has_more = input("Do you have remaining/unentered units left this term? (Y/N): ").strip().upper()
    if has_more == 'Y':
        while True:
            try:
                remaining_units = float(input("Enter the total number of remaining units/credits: "))
                if remaining_units > 0:
                    break
                print("Units must be greater than 0.")
            except ValueError:
                print("Invalid input. Please enter a valid number.")
        
        total_term_units = total_units + remaining_units
        print("\n------------------------- FORECAST -------------------------")
        
        # Checking President's Lister Goal
        target_total_weighted = 1.45 * total_term_units
        needed_weighted_sum = target_total_weighted - total_weighted_grades
        needed_avg_grade = needed_weighted_sum / remaining_units
        
        max_possible_weighted = total_weighted_grades + (1.00 * remaining_units)
        max_possible_gwa = max_possible_weighted / total_term_units
        
        if is_president_lister:
            print("To reach President's List: You have already achieved this status for your graded courses!")
            print("You need an average grade of " + str(round(needed_avg_grade, 2)) + " or better across remaining units.")
        elif needed_avg_grade < 1.00:
            print("To reach President's List: Better luck next time! It is impossible for you to join the Lister.")
            print("Your highest achievable GWA with remaining units is " + str(round(max_possible_gwa, 2)))
        else:
            print("To reach President's List:")
            print("You need an average grade of " + str(round(needed_avg_grade, 2)) + " or better across remaining units.")
            print("BUT remember your individual grades must not drop below 2.40.")
            
        # Checking Dean's Lister Goal
        target_total_weighted = 1.75 * total_term_units
        needed_weighted_sum = target_total_weighted - total_weighted_grades
        needed_avg_grade = needed_weighted_sum / remaining_units
        
        if is_president_lister or is_dean_lister:
            print("To reach Dean's List: Congratulations! Your possible GWA is within the targeted threshold.")
            print("BUT remember your individual grades must not drop below 2.40.")
        elif needed_avg_grade < 1.00:
            print("To reach Dean's List: Better luck next time! It is impossible for you to join the Lister.")
        elif needed_avg_grade > 2.40:
            print("To reach Dean's List: You need an average of 2.40 or higher,")
            print("BUT remember your individual grades must not drop below 2.40.")
        else:
            print("To reach Dean's List:")
            print("You need an average grade of " + str(round(needed_avg_grade, 2)) + " or better across remaining units.")
            
        print("------------------------------------------------------------")
    else:
        if current_gwa > 1.45:
            points_missing = current_gwa - 1.45
            print("You missed the President's List cutoff by " + str(round(points_missing, 2)) + " grade points.")

# Run the program
gwa_calculator()

#To continue loop
while True:
    y = (input ('Rerun program? (Y/N)')).upper()
    if y == 'Y':
            gwa_calculator()
    else:
            break

print ('Thanks for using my program!')
```
